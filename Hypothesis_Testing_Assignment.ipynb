2.  Imagine you are a data scientist at a company, and you want to analyze whether there is a significant difference in the generated by two sales teams: and . You collected the following data for their weekly sales revenue
(in thousands of dollars) over the past few weeks:
(in thousands of dollars): 12, 15, 14, 10, 13, 11, 12, 14, 16, 13
(in thousands of dollars): 14, 17, 16, 15, 18, 16, 15, 17, 19, 16
Hypothesis:
There is no significant difference in the average sales revenue between Team A and Team B.
There is a significant difference in the average sales revenue between Team A and Team B.
Perform an to test whether there is a significant difference in the sales revenue between Team A and Team B. Use a significance level of  0.5.



```python
from scipy import stats
import pandas as pd
import statsmodels.api as sm
TEAM_A=np.array([ 12, 15, 14, 10, 13, 11, 12, 14, 16, 13])
TEAM_B=np.array([ 14, 17, 16, 15, 18, 16, 15, 17, 19, 16])
NullH="There is no significant difference in the average sales revenue between Team A and Team B."
AlterH="There is a significant difference in the average sales revenue between Team A and Team B."
```

```python
# T TEST SAMPLE SIZE <30
test_statistics,p_value=stats.ttest_rel(TEAM_A,TEAM_B)#RELATED SAMPLES(DATA)
print("Test Statistics = ",test_statistics)
print("P Value= ",p_value)

#stats.ttest_rel(TEAM_A, TEAM_B) is a function from the scipy.stats module that performs a paired two-sample t-test
#- t-statistic: This is a measure of the difference between the means of the two samples.
#- p-value: This is the probability of observing the t-statistic (or a more extreme value) assuming that the null hypothesis is true.
#The null hypothesis is that the means of the two samples are equal.
```

```python
LOS_alpha=0.05 #5% LEVEL OF SIGNIFICANCE -PROBABILITY OF REJECTING NULL HYPOTHESIS .[CONFIDANCE 95% CHANCE OF ACCEPT H0,5%CHANCE OF  REJECT H0]
#p valueis 1.5890484139132378e-05
if  p_value < LOS_alpha:
  decision="Reject Null Hypothesis "
else:
  decision="Accept Null Hypothesis "
print(decision)
```

```python
if decision=="Reject Null Hypothesis ":
  print(AlterH)
else:
  print(NullH)
```

```python
print("**********TEST DATA--Weekly sales revenue************")
print("SALES DATA IN TEAM_A   =",TEAM_A)
print("SALES DATA  IN TEAM_B    =",TEAM_B)
print("NULL HYPOTHESIS   =",NullH)
print("ALTERNATE HYPOTHESIS  =",AlterH)
print("__________________________________\n**************HYPOTHESIS TEST RESULT****************************\n")
print("Test Statistics = ",test_statistics)
print("P Value= ",p_value)
print("Level of Significance ,alpha = ",LOS_alpha)
print("Conclusion  =",AlterH)


```

You are a data analyst working for a retail company. The company wants to analyze whether the (Cash or Card) is associated with the . The company has observed the following data:Observed Frequency Table (Payment Method vs. Age Group):

data : https://docs.google.com/spreadsheets/d/1XJmtomJk5ZtqDbY_IzNmZx-rtlKbd78QrgC-hiS80os/edit?usp=sharing
Hypothesis:
There is no association between the payment method and the customer's age group. The two variables are independent.
There is an association between the payment method and the customer's age group. The two variables are dependent.
Perform a to test whether there is an association between the payment method and the customer's age group.

```python
dataset=pd.read_csv("/AGEGROUP.csv")
dataset
```

```python
Null_H0="There is no association between the payment method and the customer's age group. The two variables are independent."
Alter_H1="There is an association between the payment method and the customer's age group. The two variables are dependent."
```

```python
#The chi-square test helps you to solve the problem in feature selection by testing the relationship between the features.
```

```python
contingency_table = dataset[['Cash', 'Card']].T#transpose
contingency_table
```

```python
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)
print('Chi-Square Statistic:', chi2_stat)
print('p-value:', p_val)
print('Degrees of Freedom:', dof)
```

```python
if p_val < 0.05:
    print('Reject the null hypothesis. There is a statistically significant association between the payment method and the customer\'s age group.')
else:
    print('Fail to reject the null hypothesis. There is no statistically significant association between the payment method and the customer\'s age group.')
```

Therefore, based on the Chi-Square Test of Independence, we can conclude that there is no statistically significant association between the payment method and the customer's age group.
